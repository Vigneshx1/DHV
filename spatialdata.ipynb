```python
import folium
mymap = folium.Map(location=[0, 0], zoom_start=2)
folium.Marker(location=[14.3134,  76.6510], popup='Challakere', icon=folium.Icon(color='blue')).add_to(mymap)
folium.Marker(location=[13.2678,76.4490], popup='Tiptur', icon=folium.Icon(color='green')).add_to(mymap)
folium.Marker(location=[131533.5,762842.0], popup='Tokyo', icon=folium.Icon(color='red')).add_to(mymap)
folium.Marker([12.9958, 77.6039], popup='Pointer 1', icon=folium.Icon(color='red')).add_to(mymap)
folium.Marker([ 12.9741, 77.6081], popup='MG road', icon=folium.Icon(color='black')).add_to(mymap)

mymap
```

```python
import folium

cities = [
    {"name": "pu", "location": [13.1718, 77.5362], "symbol": "university"},
    {"name": "house", "location": [12.9767, 77.5713], "symbol": "hospital]"},
    {"name": "chinnaswamy stadium", "location": [12.9788, 77.5997], "symbol": "stadium"},

]

mymap = folium.Map(location=[0, 0], zoom_start=2)

for city in cities:
     folium.Marker(location=city["location"], popup=city["name"], icon=folium.Icon(icon=city["symbol"], prefix='fa')).add_to(mymap)
mymap

```
